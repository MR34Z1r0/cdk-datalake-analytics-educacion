```python
import boto3
from boto3.dynamodb.conditions import Attr

# Configuración
region = 'us-east-1'
boto3.setup_default_session(profile_name='prd-valorx-admin', region_name=region)
dynamodb_client = boto3.client('dynamodb')
dynamodb_resource = boto3.resource('dynamodb')  # Para usar condiciones fácilmente
DYNAMODB_DATABASE_NAME = 'sofia-dev-datalake-credentials-ddb'
table = dynamodb_resource.Table(DYNAMODB_DATABASE_NAME)

# Valores a filtrar
data_source_filter = ['apdayc']
team_filter = 'sofia'

# Escaneo con filtro
response = table.scan(
    FilterExpression=Attr('DATA_SOURCE').is_in(data_source_filter) & 
                     Attr('TEAM').eq(team_filter)
)

response.get('Items', [])
```
